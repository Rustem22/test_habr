```python
import numpy as np
from scipy import stats

# Создание словаря для хранения результатов
anomalies = {}

# Перебор каждой специальности и каждого грейда
for spec in specializations:
    for grade in grades:
        # Фильтрация данных для текущей специальности и грейда
        filtered_data = data[(data['Specialization'] == spec) & (data['Grade'] == grade)]
        
        # Рассчитываем среднее значение и стандартное отклонение зарплаты
        mean_salary = filtered_data['Salary'].mean()
        std_salary = filtered_data['Salary'].std()
        
        # Рассчитываем 95% доверительный интервал
        confidence_interval = stats.norm.interval(0.95, loc=mean_salary, scale=std_salary / np.sqrt(len(filtered_data)))
        
        # Идентифицируем аномальные зарплаты
        anomalies_count = filtered_data[(filtered_data['Salary'] < confidence_interval[0]) | (filtered_data['Salary'] > confidence_interval[1])].shape[0]
        
        # Добавляем результаты в словарь
        anomalies[(spec, grade)] = anomalies_count

# Вывод результатов
for key, value in anomalies.items():
    print(f"Количество аномалий для {key[0]} {key[1]}: {value}")
```

```python
# Определение границ аномалий
# Например, используем межквартильный метод
Q1 = data['Salary'].quantile(0.25)
Q3 = data['Salary'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Определение индексов строк с аномальными зарплатами
anomalies_indices = data[(data['Salary'] < lower_bound) | (data['Salary'] > upper_bound)].index
print(anomalies_indices)
```

```python
# Определение аномальных значений на основе интерквартильного размаха #ЧТО ЭТО показывает?
Q1 = data['Salary'].quantile(0.25)
Q3 = data['Salary'].quantile(0.75)
IQR = Q3 - Q1
anomalies_iqr = data[(data['Salary'] < Q1 - 1.5 * IQR) | (data['Salary'] > Q3 + 1.5 * IQR)]

print("Аномальные значения на основе интерквартильного размаха:")
print(anomalies_iqr)
```

```python
# Создаем пустой список для индексов аномальных значений
all_anomalies_indices = []

# Проходимся по каждой специальности
for spec in data['Specialization'].unique():
    # Проходимся по каждому грейду
    for grade in data['Grade'].unique():
        # Определяем данные только для данной специальности и грейда
        filtered_data = data[(data['Specialization'] == spec) & (data['Grade'] == grade)]
        
        # Определение границ аномалий на основе интерквартильного размаха
        Q1 = filtered_data['Salary'].quantile(0.25)
        Q3 = filtered_data['Salary'].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Определение индексов строк с аномальными зарплатами
        anomalies_indices = filtered_data[(filtered_data['Salary'] < lower_bound) | (filtered_data['Salary'] > upper_bound)].index
        
        # Добавляем найденные индексы в общий список аномалий
        all_anomalies_indices.extend(anomalies_indices)

        # Определение индексов строк с зарплатами ниже первого квартиля
        Q1 = filtered_data['Salary'].quantile(0.25)
        lower_bound_anomalies_indices = filtered_data[filtered_data['Salary'] < Q1].index
        
        # Добавляем найденные индексы в общий список аномалий
        all_anomalies_indices.extend(lower_bound_anomalies_indices)

# Преобразуем список в множество, чтобы избежать повторений
all_anomalies_indices = set(all_anomalies_indices)

# Выводим список всех индексов аномальных значений
print("Индексы аномальных значений:", all_anomalies_indices)
```

```python
# Подсчет количества уникальных индексов аномальных значений
total_anomalies = len(all_anomalies_indices)

print("Всего аномальных значений:", total_anomalies)
```

```python
# Удаление строк с аномальными индексами
cleaned_data = data.drop(index=all_anomalies_indices)

# Проверяем размерность очищенных данных
print("Размерность очищенных данных после удаления аномалий:", cleaned_data.shape)
```
