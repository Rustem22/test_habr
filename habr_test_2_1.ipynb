```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

# Список грейдов
grades = ['Teamlead', 'Senior', 'Middle', 'Junior', 'Intern']

# Создаем графики для каждого грейда
for grade in grades:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Salary', y='Specialization', data=data[data['Grade'] == grade])
    plt.title(f'Зарплаты {grade.lower()} по направлениям')
    plt.xlabel('Зарплата')
    plt.ylabel('Направления')

    # Получаем квартили
    q1 = data[data['Grade'] == grade]['Salary'].quantile(0.25)
    q3 = data[data['Grade'] == grade]['Salary'].quantile(0.75)

    # Вычисляем межквартильный размах
    iqr = q3 - q1

    # Вычисляем нижнюю и верхнюю границы
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Выводим границы
    print(f"Для грейда {grade}, нижняя граница: {lower_bound}, верхняя граница: {upper_bound}")
    
    # Получаем элементы графика
    for line in plt.gca().lines:
        if len(line.get_xdata()) > 0 and len(line.get_ydata()) > 0:
            # Получаем координаты точек
            x_points = line.get_xdata()
            y_points = line.get_ydata()

            # Выводим значения точек
            for x, y in zip(x_points, y_points):
                plt.text(x, y, f'({x:.2f}, {y:.2f})', fontsize=8, ha='center', va='bottom', color='red')

    plt.show()
    
```

```python
# Список грейдов
grades = ['Teamlead', 'Senior', 'Middle', 'Junior', 'Intern']

# Создаем пустой DataFrame для хранения отфильтрованных данных
filtered_data_1 = pd.DataFrame()

# Применяем фильтрацию для каждого грейда
for grade in grades:
    # Вычисляем квартили для данного грейда
    q1 = data[data['Grade'] == grade]['Salary'].quantile(0.25)
    q3 = data[data['Grade'] == grade]['Salary'].quantile(0.75)
    
    # Вычисляем межквартильный размах для данного грейда
    iqr = q3 - q1
    
    # Вычисляем нижнюю и верхнюю границы для данного грейда
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Фильтруем данные для данного грейда и добавляем в отфильтрованный DataFrame
    filtered_data_1 = pd.concat([filtered_data_1, data[(data['Grade'] == grade) & (data['Salary'] >= lower_bound) & (data['Salary'] <= upper_bound)]])
```
